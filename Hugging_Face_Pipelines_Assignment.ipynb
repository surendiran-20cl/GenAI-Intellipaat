1. Sentiment Analysis

Q1. Use the Hugging Face pipeline for sentiment analysis to classify the sentiment of the following sentences. Provide the output and explain the model’s predictions in your own words:

"I absolutely loved the new iPhone!"


"The food was cold and took forever to arrive."


```python
!pip install transformers
```

```python
from transformers import pipeline

# Load sentiment-analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Input sentences
sentences = [
    "I absolutely loved the new iPhone!",
    "The food was cold and took forever to arrive."
]

# Analyze sentiment
for sentence in sentences:
    result = sentiment_analyzer(sentence)[0]
    print(f"Sentence: {sentence}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.2f}\n")
```

Explanation:
This uses a pre-trained model to classify the sentence as either positive or negative with a confidence score.

2. Named Entity Recognition (NER)

Q2. Use a NER pipeline to extract and categorize entities from this text:

 "Barack Obama was born in Hawaii and served as the 44th President of the United States. He studied at Harvard Law School."
 List all entities along with their entity type (e.g., PERSON, LOCATION).


```python
# Load NER pipeline
ner = pipeline("ner", grouped_entities=True)

# Input text
text = "Barack Obama was born in Hawaii and served as the 44th President of the United States. He studied at Harvard Law School."

# Perform NER
entities = ner(text)
for entity in entities:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Confidence: {entity['score']:.2f}")
```

Explanation:
This identifies entities like people (PERSON), places (LOCATION), and organizations (ORG).

3. Text Summarization

Q3. Summarize the following passage using the summarization pipeline (you can choose your own model or use default):

"Artificial Intelligence (AI) is transforming the way businesses operate, from automating customer service to predicting user behavior. Many industries, including healthcare, finance, and logistics, are rapidly adopting AI technologies to improve efficiency and decision-making."

 Explain how summarization models decide what to keep and what to discard.


```python
# Load summarization pipeline
summarizer = pipeline("summarization")

# Long text to summarize
text = """
Artificial Intelligence (AI) is transforming the way businesses operate, from automating customer service to predicting user behavior.
Many industries, including healthcare, finance, and logistics, are rapidly adopting AI technologies to improve efficiency and decision-making.
"""

# Generate summary
summary = summarizer(text, max_length=50, min_length=20, do_sample=False)
print("Summary:", summary[0]['summary_text'])
```

Explanation:
Summarization models keep key information (main topics, results) and discard repetitive or non-essential details.

5. Question Answering

Q5. Given the context below, use the question-answering pipeline to answer the following questions:

 Context: "Mount Everest is the tallest mountain in the world. It is part of the Himalayas and lies on the border between Nepal and China."

 Questions:

What is the tallest mountain in the world?


Where is Mount Everest located?


```python
# Load QA pipeline
qa = pipeline("question-answering")

context = "Mount Everest is the tallest mountain in the world. It is part of the Himalayas and lies on the border between Nepal and China."

questions = [
    "What is the tallest mountain in the world?",
    "Where is Mount Everest located?"
]

for question in questions:
    result = qa(question=question, context=context)
    print(f"Q: {question}")
    print(f"A: {result['answer']}\n")
```

6. Text Classification (Custom Topic Detection)

Q6. Use the text-classification pipeline with a model trained for topic or emotion detection. Classify the text below and explain your result:

 "The match was so intense, I couldn't stop cheering!"

```python
# Load a topic/emotion detection model
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")

text = "The match was so intense, I couldn't stop cheering!"
result = classifier(text)[0]
print(f"Emotion: {result['label']}, Confidence: {result['score']:.2f}")
```

Sadness

Surprise

Useful for analyzing user reactions.

8. Zero-Shot Classification

Q8. Use the zero-shot-classification pipeline to classify the sentence below into one of these categories: ["education", "technology", "sports", "health"].

 "AI models like ChatGPT are changing the way students learn and interact with content."


```python
# What it does: Classifies text into labels even if the model hasn’t seen those labels before.
zero_shot = pipeline("zero-shot-classification")

text = "AI models like ChatGPT are changing the way students learn and interact with content."
labels = ["education", "technology", "sports", "health"]

result = zero_shot(text, candidate_labels=labels)
print("Predicted Label:", result['labels'][0])
```

 Explanation:
Zero-shot learning lets you give any custom label list and the model will match the text to the most suitable one.

10. Model Comparison

Q10. Choose two different models for the same task (e.g., summarization or text generation) using Hugging Face’s pipeline. Compare their outputs on the same input and explain the difference in:

Output quality


Length


Fluency


Relevance to input




```python
from transformers import pipeline

# GPT-2 model
generator_1 = pipeline("text-generation", model="gpt2")
output1 = generator_1("The future of AI is", max_length=50)[0]['generated_text']

# DistilGPT-2 model
generator_2 = pipeline("text-generation", model="distilgpt2")
output2 = generator_2("The future of AI is", max_length=50)[0]['generated_text']

print("Model 1 Output (gpt2):\n", output1)
print("\nModel 2 Output (distilgpt2):\n", output2)
```

Comparison Criteria:

Fluency – Does it read smoothly?

Relevance – Does it stay on-topic?

Creativity – Any unique or surprising ideas?

Length/Depth – How long and informative is it?
