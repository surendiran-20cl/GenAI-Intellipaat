```python
# Chat Function
def chat():
    print("Hello! I'm your chatbot. Type 'exit' to quit the program.\n")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Bot: Goodbye!, See you again")
            break

        prompt = f"[INST] {user_input} [/INST]"
        response = Chatbot(prompt, max_new_tokens=150)[0]['generated_text']
        reply = response.split('[/INST]')[-1].strip()
        print("Bot:", reply)
```

```python
# Falcon Refined Web (1B parameter)
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python

import transformers
import torch

model = "tiiuae/falcon-rw-1b"


```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline( #chatbot ~ pipeline
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
sequences = pipeline(
   "Girafatron is obsessed with giraffes, the most glorious animal on the face of this Earth. Giraftron believes all other animals are irrelevant when compared to the glorious majesty of the giraffe.\nDaniel: Hello, Girafatron!\nGirafatron:",
    max_length=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
)
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

```python
def chat(user_input):
    prompt = f"Answer the following question:\n\n{user_input}"
    response = pipeline(prompt, max_new_tokens=100)
    return response[0]['generated_text']
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Bot: Goodbye!, See you again")
        break
    reply = chat(user_input)
    print("Bot:", reply)
```
